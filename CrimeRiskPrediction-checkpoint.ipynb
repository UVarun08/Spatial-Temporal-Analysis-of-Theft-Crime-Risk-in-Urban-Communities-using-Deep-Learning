```python
#function to calculate RMSE, MAPE and R2Square from predicted and true values
def calculateMetrics(algorithm, predict, test_labels):
    mse_error = mean_squared_error(test_labels, predict)
    rmse_error = sqrt(mse_error)
    mape_error = mean_squared_error(test_labels, predict)
    predict = predict.reshape(-1, 1)
    predict = scaler1.inverse_transform(predict)
    test_label = scaler1.inverse_transform(test_labels)
    predict = predict.ravel()
    test_label = test_label.ravel()
    square_error = r2_score(test_label, predict)
    rsquare.append(square_error)
    rmse.append(rmse_error)
    mape.append(mape_error)
    print()
    print(algorithm+" MAPE : "+str(mape_error))
    print(algorithm+" RMSE : "+str(rmse_error))
    print(algorithm+" R2 : "+str(square_error))
    print()
    for i in range(0, 10):
        print("True Crime Rate : "+str(test_label[i])+" Predicted Crime Rate : "+str(predict[i]))
    plt.figure(figsize=(5,3))
    plt.plot(test_label[0:100], color = 'red', label = 'True Crime Rate')
    plt.plot(predict[0:100], color = 'green', label = 'Predicted Crime Rate')
    plt.title(algorithm+' Crime Rate Prediction Graph')
    plt.xlabel('Test Drime Data')
    plt.ylabel('Crime Rate Prediction')
    plt.legend()
    plt.show() 
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt
from torch.utils.data import DataLoader, TensorDataset
```

```python
def build_adjacency_matrix(X, num_nodes=10):
    """
    Constructs a normalised adjacency matrix A_hat = D^{-1/2} A D^{-1/2}
    from the Community Area column in X.
    
    - Extracts unique community IDs from the data.
    - Connects communities that are numerically adjacent (|i - j| <= 1)
      simulating the topological neighbour relation in the paper.
    - Adds self-loops (A = A + I) so each node attends to itself.
    - Normalises using symmetric degree normalisation.
    
    Args:
        X        : raw feature matrix (n_samples, n_features)
        num_nodes: how many community nodes to build the graph over
    Returns:
        A_hat (num_nodes, num_nodes) normalised adjacency — numpy array
    """
    community_col = 0  # Community Area is the first feature after preprocessing
 
    # Get the top-`num_nodes` most frequent communities
    communities = X[:, community_col]
    unique_communities = np.unique(communities)[:num_nodes]
    n = len(unique_communities)
 
    # Build raw adjacency: connect communities that are spatial neighbours
    A = np.zeros((n, n), dtype=np.float32)
    for i in range(n):
        for j in range(n):
            # communities are adjacent if their IDs differ by at most 1
            if abs(unique_communities[i] - unique_communities[j]) <= 1:
                A[i, j] = 1.0
 
    # Add self-loops: A_hat = A + I  (paper uses this standard GCN trick)
    A = A + np.eye(n, dtype=np.float32)
 
    # Symmetric normalisation: D^{-1/2} A D^{-1/2}
    degree = A.sum(axis=1)
    D_inv_sqrt = np.diag(1.0 / np.sqrt(degree + 1e-8))
    A_hat = D_inv_sqrt @ A @ D_inv_sqrt
 
    return A_hat.astype(np.float32)
```

```python
class GraphConvLayer(nn.Module):
    """
    Single graph convolution layer implementing Eq.(5) from the paper.
    f(X, A) = ReLU( A_hat @ X @ W + b )
    where A_hat is the pre-normalised adjacency matrix.
    """
    def __init__(self, in_features, out_features):
        super(GraphConvLayer, self).__init__()
        self.W = nn.Linear(in_features, out_features, bias=True)
        nn.init.xavier_uniform_(self.W.weight)
 
    def forward(self, X, A_hat):
        """
        X     : (batch, num_nodes, in_features)
        A_hat : (num_nodes, num_nodes)
        """
        # Graph diffusion: propagate features across neighbours
        AX = torch.matmul(A_hat, X)           # (batch, num_nodes, in_features)
        out = self.W(AX)                       # (batch, num_nodes, out_features)
        return F.relu(out)
```

```python
class ResidualBlock(nn.Module):
    """
    ResNet-style residual unit wrapping two graph conv layers.
    Implements the residual units described in Section IV-A-1.
    
    Output = GraphConv2( GraphConv1(X) ) + X   (skip connection)
    A projection is added when channel dimensions differ.
    """
    def __init__(self, in_features, out_features, num_nodes):
        super(ResidualBlock, self).__init__()
        self.gcn1 = GraphConvLayer(in_features, out_features)
        self.gcn2 = GraphConvLayer(out_features, out_features)
        self.bn1  = nn.BatchNorm1d(num_nodes)
        self.bn2  = nn.BatchNorm1d(num_nodes)
 
        # Project residual if dimensions differ
        self.projection = None
        if in_features != out_features:
            self.projection = nn.Linear(in_features, out_features, bias=False)
 
    def forward(self, X, A_hat):
        residual = X
 
        out = self.gcn1(X, A_hat)
        out = self.bn1(out)
        out = self.gcn2(out, A_hat)
        out = self.bn2(out)
 
        if self.projection is not None:
            residual = self.projection(X)
 
        return F.relu(out + residual)    # skip connection
```

```python
class STGCN(nn.Module):
    """
    Integrated ST-GCN model combining:
    1. Spatial stream  : GraphConv + Residual blocks (captures crime
                         transition between neighbouring communities)
    2. Temporal stream : LSTM (captures within-community crime trends)
    3. Fusion layer    : concatenates both → Dense → crime count prediction
    
    Input shape  : (batch_size, num_nodes, in_features)
    Output shape : (batch_size, 1)
    """
    def __init__(self, in_features, num_nodes,
                 gcn_hidden=64, lstm_hidden=64,
                 num_residual_blocks=3, lstm_layers=2, dropout=0.2):
        super(STGCN, self).__init__()
 
        self.num_nodes  = num_nodes
        self.in_features = in_features
 
        # ── Spatial-Temporal Module ──────────────────────────────
        # Initial GCN to project input features
        self.input_gcn = GraphConvLayer(in_features, gcn_hidden)
 
        # Residual blocks — paper uses 18 units, we use num_residual_blocks
        # (scaled down to match dataset size and avoid overfitting)
        self.res_blocks = nn.ModuleList([
            ResidualBlock(gcn_hidden, gcn_hidden, num_nodes)
            for _ in range(num_residual_blocks)
        ])
 
        # Final GCN to aggregate spatial features
        self.output_gcn = GraphConvLayer(gcn_hidden, gcn_hidden // 2)
 
        # Flatten spatial output: num_nodes * (gcn_hidden//2)
        spatial_out_dim = num_nodes * (gcn_hidden // 2)
 
        # ── Temporal Module (LSTM) ───────────────────────────────
        # Input to LSTM: each node's features over time treated as sequence
        # We treat num_nodes as time steps, in_features as input_size
        self.lstm = nn.LSTM(
            input_size  = in_features,
            hidden_size = lstm_hidden,
            num_layers  = lstm_layers,
            batch_first = True,
            dropout     = dropout if lstm_layers > 1 else 0.0
        )
        temporal_out_dim = lstm_hidden
 
        # ── Fusion + Output ──────────────────────────────────────
        fused_dim = spatial_out_dim + temporal_out_dim
        self.dropout    = nn.Dropout(dropout)
        self.fc1        = nn.Linear(fused_dim, 128)
        self.fc2        = nn.Linear(128, 64)
        self.output     = nn.Linear(64, 1)
 
    def forward(self, X, A_hat):
        """
        X     : (batch, num_nodes, in_features)
        A_hat : (num_nodes, num_nodes) — pre-computed normalised adjacency
        """
        # ── Spatial stream ───────────────────────────────────────
        sp = self.input_gcn(X, A_hat)          # (batch, num_nodes, gcn_hidden)
        for res_block in self.res_blocks:
            sp = res_block(sp, A_hat)          # residual refinement
        sp = self.output_gcn(sp, A_hat)        # (batch, num_nodes, gcn_hidden//2)
        sp = sp.reshape(sp.size(0), -1)        # (batch, num_nodes * gcn_hidden//2)
 
        # ── Temporal stream ──────────────────────────────────────
        lstm_out, _ = self.lstm(X)             # (batch, num_nodes, lstm_hidden)
        tm = lstm_out[:, -1, :]                # take last time step output
 
        # ── Fusion ───────────────────────────────────────────────
        fused = torch.cat([sp, tm], dim=1)     # (batch, spatial_dim + temporal_dim)
        fused = self.dropout(fused)
        fused = F.relu(self.fc1(fused))
        fused = self.dropout(fused)
        fused = F.relu(self.fc2(fused))
        out   = self.output(fused)             # (batch, 1)
 
        return out
```

```python
NUM_NODES   = 10   # number of community nodes per graph
BATCH_SIZE  = 64
EPOCHS      = 30
LR          = 0.001
DEVICE      = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
 
print(f"Using device: {DEVICE}")
print(f"X_train shape before reshape: {X_train.shape}")
 
n_features_total = X_train.shape[1]
 
# Pad or trim features so they divide evenly into NUM_NODES
features_per_node = n_features_total // NUM_NODES
remainder         = n_features_total  % NUM_NODES
 
if remainder != 0:
    # Pad with zeros so shape is divisible
    pad_cols   = NUM_NODES - remainder
    X_train_g  = np.hstack([X_train, np.zeros((X_train.shape[0], pad_cols))])
    X_test_g   = np.hstack([X_test,  np.zeros((X_test.shape[0],  pad_cols))])
    features_per_node = (n_features_total + pad_cols) // NUM_NODES
else:
    X_train_g  = X_train.copy()
    X_test_g   = X_test.copy()
 
# Reshape: (n_samples, num_nodes, features_per_node)
X_train_graph = X_train_g.reshape(-1, NUM_NODES, features_per_node)
X_test_graph  = X_test_g.reshape(-1, NUM_NODES, features_per_node)
 
print(f"Graph input shape  — Train : {X_train_graph.shape}")
print(f"Graph input shape  — Test  : {X_test_graph.shape}")
 
# Build normalised adjacency matrix (fixed for all samples)
A_hat_np = build_adjacency_matrix(X_train, num_nodes=NUM_NODES)
A_hat    = torch.tensor(A_hat_np).to(DEVICE)
print(f"Adjacency matrix shape     : {A_hat.shape}")
 
# Convert to PyTorch tensors
X_tr = torch.tensor(X_train_graph, dtype=torch.float32)
y_tr = torch.tensor(y_train,       dtype=torch.float32)
X_te = torch.tensor(X_test_graph,  dtype=torch.float32)
y_te = torch.tensor(y_test,        dtype=torch.float32)
 
train_loader = DataLoader(TensorDataset(X_tr, y_tr),
                          batch_size=BATCH_SIZE, shuffle=True)
test_loader  = DataLoader(TensorDataset(X_te, y_te),
                          batch_size=BATCH_SIZE, shuffle=False)
```

```python
import os

WEIGHTS_PATH = "model/stgcn_proper_weights.pt"

# ==============================
# 1. MODEL CREATION
# ==============================
stgcn = STGCN(
    in_features=features_per_node,
    num_nodes=NUM_NODES,
    gcn_hidden=64,
    lstm_hidden=64,
    num_residual_blocks=3,
    lstm_layers=2,
    dropout=0.2
).to(DEVICE)

print("\nST-GCN Model Architecture:")
print(stgcn)

total_params = sum(p.numel() for p in stgcn.parameters() if p.requires_grad)
print(f"Total trainable parameters: {total_params:,}\n")

# ==============================
# 2. OPTIMIZER & LOSS
# ==============================
optimizer  = torch.optim.Adam(stgcn.parameters(), lr=LR, weight_decay=1e-4)
scheduler  = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=EPOCHS)
criterion  = nn.MSELoss()

train_losses = []
val_losses   = []
best_val_loss = float('inf')
patience_counter = 0
PATIENCE = 10

# ==============================
# 3. LOAD OR TRAIN
# ==============================
if os.path.exists(WEIGHTS_PATH):
    print("✅ Found saved model — loading weights (NO training)")

    stgcn.load_state_dict(torch.load(WEIGHTS_PATH, map_location=DEVICE))
    stgcn.eval()

else:
    print("=" * 55)
    print("🚀 Training ST-GCN Model...")
    print("=" * 55)

    for epoch in range(EPOCHS):
        # ---- TRAIN ----
        stgcn.train()
        epoch_train_loss = 0.0

        for X_batch, y_batch in train_loader:
            X_batch = X_batch.to(DEVICE)
            y_batch = y_batch.to(DEVICE)

            optimizer.zero_grad()
            pred = stgcn(X_batch, A_hat)
            loss = criterion(pred, y_batch)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(stgcn.parameters(), 1.0)
            optimizer.step()

            epoch_train_loss += loss.item()

        epoch_train_loss /= len(train_loader)

        # ---- VALIDATION ----
        stgcn.eval()
        epoch_val_loss = 0.0

        with torch.no_grad():
            for X_batch, y_batch in test_loader:
                X_batch = X_batch.to(DEVICE)
                y_batch = y_batch.to(DEVICE)

                pred = stgcn(X_batch, A_hat)
                loss = criterion(pred, y_batch)

                epoch_val_loss += loss.item()

        epoch_val_loss /= len(test_loader)
        scheduler.step()

        train_losses.append(epoch_train_loss)
        val_losses.append(epoch_val_loss)

        # ---- SAVE BEST MODEL ----
        if epoch_val_loss < best_val_loss:
            best_val_loss = epoch_val_loss
            torch.save(stgcn.state_dict(), WEIGHTS_PATH)
            patience_counter = 0
        else:
            patience_counter += 1

        print(f"Epoch [{epoch+1:02d}/{EPOCHS}]  "
              f"Train Loss: {epoch_train_loss:.6f}  "
              f"Val Loss: {epoch_val_loss:.6f}")

        if patience_counter >= PATIENCE:
            print(f"\nEarly stopping at epoch {epoch+1}")
            break

    print("✅ Training complete")

# ==============================
# 4. LOAD FINAL MODEL (ALWAYS)
# ==============================
stgcn.load_state_dict(torch.load(WEIGHTS_PATH, map_location=DEVICE))
stgcn.eval()

print("✅ Model ready (loaded best weights)")

# ==============================
# 5. OPTIONAL: PLOT LOSS
# ==============================
if len(train_losses) > 0:
    plt.figure(figsize=(7, 3))
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Val Loss')
    plt.title("ST-GCN Training & Validation Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
def calculateMetrics(algorithm, predict, test_labels):
    # --- compute errors with distinct local names ---
    _mse       = mean_squared_error(test_labels, predict)
    _rmse      = sqrt(_mse)
    _mape      = mean_squared_error(test_labels, predict)  # kept consistent with original

    predict_2d    = predict.reshape(-1, 1)
    pred_inv      = scaler1.inverse_transform(predict_2d).ravel()
    label_inv     = scaler1.inverse_transform(test_labels.reshape(-1, 1)).ravel()

    _r2 = r2_score(label_inv, pred_inv)

    # --- append to GLOBAL lists (no name clash now) ---
    rsquare.append(_r2)
    rmse.append(_rmse)
    mape.append(_mape)

    print()
    print(algorithm + " MAPE : " + str(_mape))
    print(algorithm + " RMSE : " + str(_rmse))
    print(algorithm + " R2   : " + str(_r2))
    print()

    for i in range(min(10, len(label_inv))):
        print("True Crime Rate : " + str(label_inv[i]) +
              "  Predicted Crime Rate : " + str(pred_inv[i]))

    plt.figure(figsize=(5, 3))
    plt.plot(label_inv[:100], color='red',   label='True Crime Rate')
    plt.plot(pred_inv[:100],  color='green', label='Predicted Crime Rate')
    plt.title(algorithm + ' Crime Rate Prediction Graph')
    plt.xlabel('Test Crime Data')
    plt.ylabel('Crime Rate Prediction')
    plt.legend()
    plt.show()
```

```python
# ================================================================
# CELL 41 — Generate HTML Dashboard  (run AFTER Cell 40)
# Opens automatically in your browser from Jupyter
# ================================================================

import json, os, webbrowser, pathlib

json_data = json.dumps(results, indent=2)

html = f"""<!DOCTYPE html>
<html lang="en">
<head>
<meta charset="UTF-8"/>
<meta name="viewport" content="width=device-width,initial-scale=1.0"/>
<title>Chicago Crime Risk Dashboard</title>
<link href="https://fonts.googleapis.com/css2?family=Sora:wght@300;400;500;600;700&family=JetBrains+Mono:wght@400;500&display=swap" rel="stylesheet"/>
<style>
*,*::before,*::after{{box-sizing:border-box;margin:0;padding:0}}

:root{{
  --bg:#07090d;--surface:#0e1117;--card:#121924;--border:#1c2535;
  --text:#d0dae8;--muted:#4d5f75;--accent:#3b82f6;--accent-glow:rgba(59,130,246,.15);
  --low:#22c55e;--med:#f59e0b;--high:#ef4444;
  --low-bg:rgba(34,197,94,.1);--med-bg:rgba(245,158,11,.1);--high-bg:rgba(239,68,68,.1);
  --track:rgba(255,255,255,.06);--tag-bg:rgba(255,255,255,.04);
  --shadow:rgba(0,0,0,.5);
}}
[data-theme=light]{{
  --bg:#f4f7fb;--surface:#ffffff;--card:#ffffff;--border:#e2e8f0;
  --text:#1e2d3d;--muted:#94a3b8;--accent:#2563eb;--accent-glow:rgba(37,99,235,.08);
  --low:#16a34a;--med:#d97706;--high:#dc2626;
  --low-bg:rgba(22,163,74,.08);--med-bg:rgba(217,119,6,.08);--high-bg:rgba(220,38,38,.08);
  --track:rgba(0,0,0,.06);--tag-bg:rgba(0,0,0,.04);
  --shadow:rgba(0,0,0,.08);
}}

body{{
  background:var(--bg);color:var(--text);
  font-family:'Sora',sans-serif;min-height:100vh;
  transition:background .25s,color .25s
}}

/* subtle dot grid */
body::before{{
  content:'';position:fixed;inset:0;pointer-events:none;z-index:0;
  background-image:radial-gradient(rgba(59,130,246,.07) 1px,transparent 1px);
  background-size:32px 32px
}}

/* ─── HEADER ─── */
header{{
  position:relative;z-index:10;
  padding:32px 48px 24px;
  border-bottom:1px solid var(--border);
  display:flex;align-items:center;justify-content:space-between;gap:20px;flex-wrap:wrap
}}
.brand{{display:flex;flex-direction:column;gap:4px}}
.brand-top{{display:flex;align-items:center;gap:12px}}
.brand-icon{{
  width:36px;height:36px;border-radius:9px;
  background:var(--accent);display:flex;align-items:center;
  justify-content:center;font-size:17px;flex-shrink:0
}}
.brand h1{{
  font-size:1.35rem;font-weight:700;color:var(--text);letter-spacing:-.01em
}}
.brand-sub{{
  font-size:.72rem;color:var(--muted);letter-spacing:.04em;margin-left:48px
}}
.header-right{{display:flex;align-items:center;gap:10px}}
.model-badge{{
  padding:6px 14px;border-radius:7px;border:1px solid var(--border);
  background:var(--surface);font-family:'JetBrains Mono',monospace;
  font-size:.7rem;color:var(--accent);letter-spacing:.04em
}}
.theme-btn{{
  display:flex;align-items:center;gap:6px;
  padding:7px 14px;border-radius:7px;border:1px solid var(--border);
  background:var(--surface);color:var(--muted);cursor:pointer;
  font-family:'Sora',sans-serif;font-size:.73rem;font-weight:500;
  transition:all .2s
}}
.theme-btn:hover{{border-color:var(--accent);color:var(--accent)}}

/* ─── STAT BAR ─── */
.stat-bar{{
  position:relative;z-index:10;
  display:grid;grid-template-columns:repeat(5,1fr);
  border-bottom:1px solid var(--border)
}}
.stat{{padding:16px 24px;border-right:1px solid var(--border)}}
.stat:last-child{{border-right:none}}
.stat-v{{
  font-family:'JetBrains Mono',monospace;
  font-size:1.75rem;font-weight:500;color:var(--text);line-height:1
}}
.stat-v.low{{color:var(--low)}}.stat-v.med{{color:var(--med)}}.stat-v.high{{color:var(--high)}}
.stat-k{{font-size:.62rem;color:var(--muted);text-transform:uppercase;letter-spacing:.07em;margin-top:4px}}

/* ─── CHART ROW ─── */
.chart-row{{
  position:relative;z-index:10;
  display:grid;grid-template-columns:1fr 1fr 1fr;
  border-bottom:1px solid var(--border)
}}
.chart-box{{padding:22px 28px;border-right:1px solid var(--border)}}
.chart-box:last-child{{border-right:none}}
.chart-label{{font-size:.63rem;color:var(--muted);text-transform:uppercase;letter-spacing:.08em;margin-bottom:14px}}

/* donut */
.donut-wrap{{display:flex;align-items:center;gap:18px}}
svg.donut{{width:72px;height:72px;flex-shrink:0}}
.legend{{display:flex;flex-direction:column;gap:6px}}
.leg-row{{display:flex;align-items:center;gap:7px;font-size:.72rem}}
.leg-dot{{width:7px;height:7px;border-radius:50%;flex-shrink:0}}

/* time bars */
.tbars{{display:flex;flex-direction:column;gap:6px}}
.tbar-row{{display:flex;align-items:center;gap:8px;font-size:.68rem}}
.tbar-lbl{{width:72px;color:var(--muted);flex-shrink:0}}
.tbar-track{{flex:1;height:6px;background:var(--track);border-radius:3px;overflow:hidden}}
.tbar-fill{{height:100%;border-radius:3px;background:var(--accent);opacity:.75}}
.tbar-n{{width:20px;text-align:right;color:var(--text)}}

/* histogram */
.hist{{display:flex;align-items:flex-end;gap:3px;height:54px}}
.hist-col{{flex:1;border-radius:2px 2px 0 0;min-height:2px;cursor:default;transition:opacity .2s}}
.hist-col:hover{{opacity:1!important}}

/* ─── CONTROLS ─── */
.controls{{
  position:relative;z-index:10;padding:12px 48px;
  border-bottom:1px solid var(--border);
  display:flex;gap:8px;align-items:center;flex-wrap:wrap
}}
.ctrl-label{{font-size:.63rem;color:var(--muted);text-transform:uppercase;letter-spacing:.07em;margin-right:2px}}
.chip{{
  padding:5px 13px;border-radius:20px;font-size:.72rem;font-weight:500;cursor:pointer;
  border:1px solid var(--border);background:var(--surface);color:var(--muted);
  transition:all .17s
}}
.chip:hover{{border-color:var(--accent);color:var(--accent)}}
.chip.sel-all{{border-color:var(--accent);color:var(--accent);background:var(--accent-glow)}}
.chip.sel-low{{border-color:var(--low);color:var(--low);background:var(--low-bg)}}
.chip.sel-med{{border-color:var(--med);color:var(--med);background:var(--med-bg)}}
.chip.sel-high{{border-color:var(--high);color:var(--high);background:var(--high-bg)}}
.ctrl-right{{margin-left:auto;display:flex;gap:8px;align-items:center}}
select.sort-dd,input.search{{
  background:var(--surface);border:1px solid var(--border);border-radius:7px;
  padding:6px 11px;color:var(--text);font-family:'Sora',sans-serif;
  font-size:.72rem;outline:none;transition:border-color .2s;cursor:pointer
}}
select.sort-dd:focus,input.search:focus{{border-color:var(--accent)}}
input.search{{width:210px;cursor:text}}
input.search::placeholder{{color:var(--muted)}}

/* ─── GRID ─── */
.grid{{
  position:relative;z-index:10;
  padding:24px 48px 60px;
  display:grid;
  grid-template-columns:repeat(auto-fill,minmax(310px,1fr));
  gap:14px
}}
.no-res{{
  grid-column:1/-1;text-align:center;padding:80px;
  color:var(--muted);font-size:.88rem
}}

/* ─── CARD ─── */
.card{{
  background:var(--card);border:1px solid var(--border);border-radius:12px;
  overflow:hidden;cursor:default;
  transition:transform .2s,border-color .2s,box-shadow .2s;
  animation:rise .3s ease both
}}
.card:hover{{
  transform:translateY(-3px);
  border-color:var(--accent);
  box-shadow:0 10px 28px var(--shadow)
}}
@keyframes rise{{from{{opacity:0;transform:translateY(12px)}}to{{opacity:1;transform:translateY(0)}}}}

.card-top{{
  padding:11px 15px 10px;border-bottom:1px solid var(--border);
  display:flex;justify-content:space-between;align-items:center
}}
.card-num{{font-family:'JetBrains Mono',monospace;font-size:.6rem;color:var(--muted)}}
.pill{{
  font-size:.6rem;font-weight:600;padding:3px 9px;border-radius:20px;
  text-transform:uppercase;letter-spacing:.07em
}}
.p-LOW   {{background:var(--low-bg);color:var(--low);border:1px solid var(--low)}}
.p-MEDIUM{{background:var(--med-bg);color:var(--med);border:1px solid var(--med)}}
.p-HIGH  {{background:var(--high-bg);color:var(--high);border:1px solid var(--high)}}

.card-body{{padding:12px 15px}}
.area-name{{
  font-size:.92rem;font-weight:600;color:var(--text);
  white-space:nowrap;overflow:hidden;text-overflow:ellipsis;margin-bottom:6px
}}
.tags{{display:flex;gap:4px;flex-wrap:wrap;margin-bottom:10px}}
.tag{{
  font-size:.6rem;padding:2px 7px;border-radius:4px;
  background:var(--tag-bg);color:var(--muted)
}}
.tag.t-accent{{color:var(--accent);border:1px solid rgba(59,130,246,.2)}}
.tag.t-peak{{background:rgba(239,68,68,.1);color:var(--high)}}

/* score bar */
.score-row{{display:flex;align-items:center;gap:8px;margin-bottom:10px}}
.sc-label{{font-size:.6rem;color:var(--muted);text-transform:uppercase;letter-spacing:.05em;width:36px;flex-shrink:0}}
.sc-track{{flex:1;height:4px;background:var(--track);border-radius:2px;overflow:hidden}}
.sc-bar{{height:100%;border-radius:2px;transition:width .5s ease}}
.b-LOW   {{background:var(--low)}}
.b-MEDIUM{{background:var(--med)}}
.b-HIGH  {{background:var(--high)}}
.sc-num{{font-family:'JetBrains Mono',monospace;font-size:.75rem;color:var(--text);width:34px;text-align:right}}

/* recommendation */
.rec{{
  margin-top:8px;padding:8px 10px;border-radius:7px;
  border-left:2px solid;font-size:.68rem;line-height:1.55
}}
.rec.r-LOW   {{background:var(--low-bg);border-color:var(--low)}}
.rec.r-MEDIUM{{background:var(--med-bg);border-color:var(--med)}}
.rec.r-HIGH  {{background:var(--high-bg);border-color:var(--high)}}
.rec-head{{font-weight:600;font-size:.62rem;text-transform:uppercase;letter-spacing:.06em;margin-bottom:3px}}
.rec-head.low{{color:var(--low)}}.rec-head.med{{color:var(--med)}}.rec-head.high{{color:var(--high)}}
.rec-body{{color:var(--text);opacity:.85}}
.rec-action{{color:var(--muted);margin-top:3px;font-size:.65rem}}

/* meta strip */
.meta{{display:flex;gap:5px;flex-wrap:wrap;margin-top:9px}}
.meta-chip{{
  font-size:.6rem;color:var(--muted);
  border:1px solid var(--border);border-radius:4px;padding:2px 6px
}}
.meta-chip.coords{{color:var(--accent);border-color:rgba(59,130,246,.2)}}

/* ─── FOOTER ─── */
footer{{
  position:relative;z-index:10;
  border-top:1px solid var(--border);
  padding:18px 48px;
  display:flex;align-items:center;justify-content:space-between;
  font-size:.63rem;color:var(--muted);letter-spacing:.03em;flex-wrap:wrap;gap:8px
}}
</style>
</head>
<body data-theme="dark">

<!-- HEADER -->
<header>
  <div class="brand">
    <div class="brand-top">
      <div class="brand-icon">🔍</div>
      <h1>Crime Risk Dashboard</h1>
    </div>
    <div class="brand-sub">Chicago · Theft Risk Prediction · Integrated LSTM + ST-GCN + GBDT</div>
  </div>
  <div class="header-right">
    <div class="model-badge">LSTM + ST-GCN + GBDT</div>
    <button class="theme-btn" onclick="toggleTheme()" id="themeBtn">
      <span id="themeIcon">☀</span>
      <span id="themeLabel">Light Mode</span>
    </button>
  </div>
</header>

<!-- STATS -->
<div class="stat-bar">
  <div class="stat"><div class="stat-v" id="sTotal">—</div><div class="stat-k">Total Records</div></div>
  <div class="stat"><div class="stat-v low" id="sLow">—</div><div class="stat-k">Low Risk</div></div>
  <div class="stat"><div class="stat-v med" id="sMed">—</div><div class="stat-k">Medium Risk</div></div>
  <div class="stat"><div class="stat-v high" id="sHigh">—</div><div class="stat-k">High Risk</div></div>
  <div class="stat"><div class="stat-v" id="sAvg">—</div><div class="stat-k">Avg Risk Score</div></div>
</div>

<!-- CHARTS -->
<div class="chart-row">
  <div class="chart-box">
    <div class="chart-label">Risk Split</div>
    <div class="donut-wrap">
      <svg class="donut" viewBox="0 0 36 36">
        <circle cx="18" cy="18" r="15.9" fill="none" id="donutBase" stroke="#1c2535" stroke-width="3"/>
        <circle id="dH" cx="18" cy="18" r="15.9" fill="none" stroke="#ef4444" stroke-width="3"
                stroke-dasharray="0 100" stroke-dashoffset="25" style="transition:stroke-dasharray .55s"/>
        <circle id="dM" cx="18" cy="18" r="15.9" fill="none" stroke="#f59e0b" stroke-width="3"
                stroke-dasharray="0 100" stroke-dashoffset="25" style="transition:stroke-dasharray .55s"/>
        <circle id="dL" cx="18" cy="18" r="15.9" fill="none" stroke="#22c55e" stroke-width="3"
                stroke-dasharray="0 100" stroke-dashoffset="25" style="transition:stroke-dasharray .55s"/>
      </svg>
      <div class="legend" id="donutLegend"></div>
    </div>
  </div>
  <div class="chart-box">
    <div class="chart-label">Crimes by Time of Day</div>
    <div class="tbars" id="timeChart"></div>
  </div>
  <div class="chart-box">
    <div class="chart-label">Score Distribution</div>
    <div class="hist" id="histChart"></div>
  </div>
</div>

<!-- CONTROLS -->
<div class="controls">
  <span class="ctrl-label">Filter</span>
  <button class="chip sel-all" onclick="setFilter('ALL',this)">All</button>
  <button class="chip"         onclick="setFilter('LOW',this)">Low Risk</button>
  <button class="chip"         onclick="setFilter('MEDIUM',this)">Medium Risk</button>
  <button class="chip"         onclick="setFilter('HIGH',this)">High Risk</button>
  <div class="ctrl-right">
    <select class="sort-dd" id="sortSel" onchange="render()">
      <option value="index">Default order</option>
      <option value="score_desc">Score: highest first</option>
      <option value="score_asc">Score: lowest first</option>
      <option value="area">Area: A–Z</option>
    </select>
    <input class="search" id="searchBox" placeholder="Search area or crime type…" oninput="render()"/>
  </div>
</div>

<!-- CARD GRID -->
<div class="grid" id="cardGrid"></div>

<!-- FOOTER -->
<footer>
  <span>Chicago Crime Risk Prediction · Integrated Model (LSTM + ST-GCN + GBDT)</span>
  <span>Trained on Chicago Crimes Dataset · 77 Urban Communities</span>
</footer>

<script>
// ── DATA ─────────────────────────────────────────────────────
const DATA = {json_data};

// ── THEME ─────────────────────────────────────────────────────
function toggleTheme() {{
  const dark = document.body.getAttribute('data-theme') === 'dark';
  document.body.setAttribute('data-theme', dark ? 'light' : 'dark');
  document.getElementById('themeIcon').textContent  = dark ? '🌙' : '☀';
  document.getElementById('themeLabel').textContent = dark ? 'Dark Mode' : 'Light Mode';
  document.getElementById('donutBase').setAttribute('stroke', dark ? '#e2e8f0' : '#1c2535');
  render();
}}

// ── FILTER ────────────────────────────────────────────────────
let activeFilter = 'ALL';
function setFilter(f, el) {{
  activeFilter = f;
  document.querySelectorAll('.chip').forEach(c => c.className = 'chip');
  el.className = 'chip ' + {{ALL:'sel-all',LOW:'sel-low',MEDIUM:'sel-med',HIGH:'sel-high'}}[f];
  render();
}}

// ── CHARTS ────────────────────────────────────────────────────
function buildCharts(data) {{
  const n     = data.length || 1;
  const low   = data.filter(d => d.risk === 'LOW').length;
  const med   = data.filter(d => d.risk === 'MEDIUM').length;
  const high  = data.filter(d => d.risk === 'HIGH').length;
  const avg   = data.length
    ? (data.reduce((s,d) => s + d.score, 0) / data.length).toFixed(1) : '—';

  document.getElementById('sTotal').textContent = data.length;
  document.getElementById('sLow').textContent   = low;
  document.getElementById('sMed').textContent   = med;
  document.getElementById('sHigh').textContent  = high;
  document.getElementById('sAvg').textContent   = avg;

  // Donut arcs
  const hp = high/n*100, mp = med/n*100, lp = low/n*100;
  let off = 25;
  function setArc(id, pct) {{
    const el = document.getElementById(id);
    el.setAttribute('stroke-dasharray', pct.toFixed(1) + ' ' + (100 - pct).toFixed(1));
    el.setAttribute('stroke-dashoffset', off);
    off -= pct;
  }}
  setArc('dH', hp); setArc('dM', mp); setArc('dL', lp);

  document.getElementById('donutLegend').innerHTML = [
    ['var(--high)', 'High',   high, hp],
    ['var(--med)',  'Medium', med,  mp],
    ['var(--low)',  'Low',    low,  lp],
  ].map(([c, l, count, pct]) =>
    `<div class="leg-row">
      <div class="leg-dot" style="background:${{c}}"></div>
      <span style="color:${{c}}">${{l}}</span>
      <span style="color:var(--muted)">${{count}} (${{pct.toFixed(0)}}%)</span>
    </div>`
  ).join('');

  // Time bars
  const times = {{'Late Night':0,'Morning':0,'Afternoon':0,'Evening':0,'Night':0}};
  data.forEach(d => {{ if (d.time_label in times) times[d.time_label]++; }});
  const maxT = Math.max(...Object.values(times), 1);
  document.getElementById('timeChart').innerHTML = Object.entries(times)
    .map(([k, v]) =>
      `<div class="tbar-row">
        <div class="tbar-lbl">${{k}}</div>
        <div class="tbar-track">
          <div class="tbar-fill" style="width:${{(v/maxT*100).toFixed(0)}}%"></div>
        </div>
        <div class="tbar-n">${{v}}</div>
      </div>`
    ).join('');

  // Histogram
  const buckets = Array(10).fill(0);
  data.forEach(d => {{ buckets[Math.min(Math.floor(d.score / 10), 9)]++; }});
  const maxB = Math.max(...buckets, 1);
  const cols  = ['#22c55e','#4ade80','#84cc16','#eab308','#f59e0b',
                  '#f97316','#ef4444','#dc2626','#b91c1c','#991b1b'];
  document.getElementById('histChart').innerHTML = buckets
    .map((b, i) =>
      `<div class="hist-col"
            style="height:${{(b/maxB*100).toFixed(0)}}%;background:${{cols[i]}};opacity:.75"
            title="Score ${{i*10}}–${{i*10+9}}: ${{b}} record(s)">
      </div>`
    ).join('');
}}

// ── RENDER CARDS ──────────────────────────────────────────────
function render() {{
  const q    = document.getElementById('searchBox').value.toLowerCase();
  const srt  = document.getElementById('sortSel').value;
  const grid = document.getElementById('cardGrid');

  let rows = DATA.filter(d => {{
    const flt = activeFilter === 'ALL' || d.risk === activeFilter;
    const srh = !q || d.area.toLowerCase().includes(q)
                   || d.crime_type.toLowerCase().includes(q);
    return flt && srh;
  }});

  if (srt === 'score_desc') rows.sort((a, b) => b.score - a.score);
  else if (srt === 'score_asc')  rows.sort((a, b) => a.score - b.score);
  else if (srt === 'area')       rows.sort((a, b) => a.area.localeCompare(b.area));

  buildCharts(rows);

  if (!rows.length) {{
    grid.innerHTML = '<div class="no-res">Nothing matches your filter or search.</div>';
    return;
  }}

  const maxScore = Math.max(...DATA.map(d => d.score), 100);

  grid.innerHTML = rows.map((d, idx) => {{
    const pct  = Math.min((d.score / maxScore) * 100, 100).toFixed(1);
    const peak = d.peak_warning
      ? '<span class="tag t-peak">⚠ Peak Hour</span>' : '';
    const hlc  = {{HIGH:'high', MEDIUM:'med', LOW:'low'}}[d.risk];

    return `<div class="card" style="animation-delay:${{idx * 20}}ms">
      <div class="card-top">
        <span class="card-num">#${{String(d.index).padStart(3,'0')}}</span>
        <span class="pill p-${{d.risk}}">${{d.risk}} RISK</span>
      </div>
      <div class="card-body">
        <div class="area-name">${{d.area}}</div>
        <div class="tags">
          <span class="tag">${{d.crime_type}}</span>
          <span class="tag">${{d.crime_cat}}</span>
          <span class="tag t-accent">${{d.time_label}}</span>
          ${{peak}}
        </div>
        <div class="score-row">
          <div class="sc-label">Score</div>
          <div class="sc-track">
            <div class="sc-bar b-${{d.risk}}" style="width:${{pct}}%"></div>
          </div>
          <div class="sc-num">${{d.score}}</div>
        </div>
        <div class="rec r-${{d.risk}}">
          <div class="rec-head ${{hlc}}">${{d.risk === 'HIGH' ? 'Action needed' : d.risk === 'MEDIUM' ? 'Keep an eye out' : 'All clear'}}</div>
          <div class="rec-body">${{d.recommendation}}</div>
          <div class="rec-action">${{d.action}}</div>
        </div>
        <div class="meta">
          <span class="meta-chip coords">📍 ${{d.lat}}, ${{d.lon}}</span>
          <span class="meta-chip">${{d.date.slice(0,10)}}</span>
        </div>
      </div>
    </div>`;
  }}).join('');
}}

render();
</script>
</body>
</html>"""

# ── Write and open ─────────────────────────────────────────────
path = "crime_risk_dashboard.html"
with open(path, "w", encoding="utf-8") as f:
    f.write(html)

print(f"✅ Dashboard saved: {os.path.abspath(path)}")
webbrowser.open(pathlib.Path(path).resolve().as_uri())
print("🌐 Dashboard opened in your browser!")
```
